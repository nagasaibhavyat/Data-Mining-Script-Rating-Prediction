```python
import pandas as pd
df_epi = pd.read_csv('episodes_final_df.csv')
```

```python
df_epi.head()
```

```python
df_epi.shape
```

```python
dc = pd.read_csv('Modified_crew.csv')
dc.head()
```

```python
grouped = dc.groupby('id').agg(lambda x: x.tolist()).reset_index()
```

```python
grouped.head()
```

```python
grouped.shape
```

```python
merged_df = pd.merge(grouped, df_epi, on='id', how='outer')
merged_df.head()
```

```python
merged_df[merged_df.isnull().any(axis=1)]
```

```python
df = merged_df.dropna(subset=['air_date'])
```

```python
df[df.isnull().any(axis=1)]
```

```python
df.to_csv('all_metadat_combined.csv')
```
