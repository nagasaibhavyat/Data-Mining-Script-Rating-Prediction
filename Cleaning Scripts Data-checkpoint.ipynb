```python
#dropping the url column as it is of no use now
df = df.drop('url',axis=1)
df.head()
```

```python
#splitting episodes and titles
df[['Episode and Sesons', 'Title']] = df["season and title"].str.split(':', 1, expand = True)
df['Title'].str.strip() #removinf whitespaces if any
df.head()
```

```python
#so there is a null value here which means that one of the scripts was not scraped. So we try oto get this data - 
#this is because this particular webpage has all the data in a pre tag 
import requests
from bs4 import BeautifulSoup

curr_episode_list=[]
response = requests.get('https://www.livesinabox.com/friends/season9/907song.htm', verify=False)
soup = BeautifulSoup(response.content, 'html.parser')
body_tag = soup.find_all('pre')
df['text script'].loc[195] = body_tag[0].get_text()[body_tag[0].get_text().find("[Scene"):]
```
